## Primera Clase - Pyplot Basico 


**Pyplot:** es la forma que tenemos para representar datos de manera grafica dentro de matplotlib

## Sublot - Manejo de Matplotlib, para sacar mas de un grafico 

Subplot permite crear gráficos dentro de una gráfica. Esto lo hace a través de una matriz de gráficos y se puede acceder a ellos a través de índices:

```python
x = np.linspace(0, 5, 11)
y = x**2 
```

```python
#  Trabajremos con 2 graficas, una un plot para ver la relacion entre (x) y (y)
# En otra veremos un histograma 

plt.subplot(1, 2, 1)
plt.plot(x,y)

#! creacion del segundo grafico


plt.subplot(1, 2, 2)
plt.hist(y)
plt.show()
```

```python
plt.subplot(1,2,1) # Grafico 1
plt.plot(x,y, 'r--') 

plt.subplot(1,2,2)  # Gráfico 2
plt.pie(y)
plt.xlabel("Cantidad de ventas ")
plt.show()
```

## Metodo Orientado a Objetos 

Tenemos distintas maneras para representar graficas dentro de matplotlib, una de ellas es **Object Oriented** 

```python
x = np.linspace(0,5,11)
y = x**2 
```

```python
# Graficando con el metodo **Orientado a Objetos**
fig = plt.figure() # Lienzo de la grafica 


# 
#! Generando las graficas 
axes = fig.add_axes([0.1,0.1,0.5,0.9])


axes.plot(x,y)
fig.show()
```

```python
# Queremos hacer algo mas especifico

ig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.5,0.9])
axes2 = fig.add_axes([0.1,0.2,0.5,0.9])

axes.plot(x,y)
axes2.plot(x,y)

fig.show()
```

## Subplots

Trabajar multiples graficas con un solo lienzo

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
# Graficando usando subplots 
fig, axes = plt.subplots()
axes.plot(x,y, "b")
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2) # 
axes[0].plot(x,y,"r")


axes[1].plot(y,x,"y")
plt.show()
```

```python
fig, axes = plt.subplots(2,4)

axes[0,0].plot(x,x**2)
axes[0,1].hist(x)
axes[0,2].pie(y)

#genera un trazo accediendo a las graficas a traves del indice de la matriz
axes[1,0].plot(x,np.cos(x))
axes[1,1].plot(x,np.sin(x), 'r')
axes[1,2].plot(x,np.tan(x), 'green')
axes[1,3].plot(x,np.cos(x)**2)

fig.tight_layout() #mejora la visualizacion de los ejes de cada grafico
```

## Leyendas, Etiquetas, titulos, tamaños 

Dar contexto a nuestras graficas

```python
x = np.linspace(0,5,20)
y = np.sin(x)
```

```python
fig, axes = plt.subplots(1,2, figsize=(10,7))   ## Creando nuestros lienzos 
axes[0].set_title("Relacion X - Y") # Generando un titulo para la primera grafica


axes[0].plot(x,y, "r", label="$Sen(x)$")
axes[0].plot(x, np.cos(y), "y", label="$Cos(x)$")
axes[0].plot(x, np.tan(y), "b", label="$tan(x)$")
axes[0].plot()
# Dando etiquetas a los ejes
axes[0].set_xlabel("Datos de X")
axes[0].set_ylabel("Sen(x)")

axes[0].legend()
axes[1].set_title("Relacion de Sen()")
axes[1].plot(y,x)
axes[1].set_xlabel("Sen(x)")
axes[1].set_ylabel("Conjunto (x)")


plt.show()
```

```python
x = np.linspace(-2, 2)


def f(x): 
    return 2 * x + 4

```

```python
fig, axes = plt.subplots(1,1, figsize=(10,7))
axes.set_title("Ejercicio #1")


axes.plot(f(x), x, "black", label="$f(x) = 2x + 4$")
axes.set_xlabel("Conjunto de Datos")
axes.set_ylabel("Funcion aplicada")


axes.plot(x, np.sin(f(x)), "b", label="$sin(f(x))$")

axes.legend()
plt.show()
```

## Bar Plot

Grafico de Barras, graficar variables categoricas, variables de texto 

```python
# Graficando datos en texto

country = ["INDIA", "JAPAN", "MEXICO", "COLOMBIA", "GERMANY"]

population = [1000, 800, 900, 1000, 300]
```

```python

plt.style.use('dark_background')
fig, axes = plt.subplots(1,1, figsize=(10,12))
axes.set_title("Poblacion de Diferentes Paises")
axes.bar(country, population, color=["red", "gray"], label="Pobleacion en escala", width=0.2)

plt.xticks(rotation=35)
plt.xlabel("Pais")
plt.legend()
plt.show()
```
