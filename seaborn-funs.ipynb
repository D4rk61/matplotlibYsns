# Conociendo la libreria **Seaborn** y sus distintas propiedades y funciones especiales

```python
import seaborn as sns
import matplotlib.pyplot as plt 
import pandas as pd
```

## Comando **Set**

```python
## Comando para definir "parametros" dentro de seaborn

sns.barplot(x=["A", "B", "C"], y=[1,2,3]) 
plt.show()
```

```python
# Usando el comanod `set`, cambiando el estilo

sns.set(style="darkgrid")
sns.barplot(x=["A", "B", "C"], y=[1,2,3]) 
plt.show()
```

```python
sns.displot(data=tips, x="total_bill", y="tip") 

plt.show()
```

## Distribuciones

Distribuciones orientadas a datos numericos

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
# Implementando grafica para trabajar con distribuciones
sns.histplot(
    data=tips,
    x="tip",
    bins=15,
    cumulative=False, # Sumando entre barras, comportamiento acumulativo
    hue="sex",
    stat="frequency"
)
plt.show()
```

```python
# Trabajando con Probabilidad

# Implementando grafica para trabajar con distribuciones
sns.histplot(
    data=tips,
    x="tip",
    bins=15,
    cumulative=False, # Sumando entre barras, comportamiento acumulativo
    hue="sex",
    stat="count",
    multiple="dodge"
)
plt.show()
```

```python
# Otro grafico "kde" - diagrama de densidad

sns.kdeplot(
    data=tips,
    x="tip",
    hue="sex",
    fill=True,
    bw_adjust=1,
)

plt.show()
```

```python
# otro grafico - Grafico escalonado de la proporcion de la frecuencia 
sns.ecdfplot(
    data=tips,
    x="tip",
    hue="sex"
)
plt.show()
```
